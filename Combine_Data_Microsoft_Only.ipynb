```python
df_msft = pandas.read_csv("./data/MSFT.csv")
df_news = pandas.read_csv("./data/uci-news-aggregator.csv")
```

Add the value if the stock rose

```python
df_msft = df_msft[(df_msft['Date'] > '2014-03-09') & (df_msft['Date'] < '2014-08-29')]

df_msft
```

```python
#t = technology
df_technology = df_news[df_news['CATEGORY'] == 't']
df_technology = df_technology[df_technology['TITLE'].str.contains('microsoft', case=False)]
sLength = len(df_technology['TIMESTAMP'])
#df_technology['today'] = Series(np.zeros(sLength), index=df_technology.index, dtype=np.int32)
#df_technology['tomorrow'] = Series(np.zeros(sLength), index=df_technology.index, dtype=np.int32)

df_technology['TIMESTAMP'] = df_technology['TIMESTAMP'].apply(lambda x: datetime.fromtimestamp(int(int(x)/1000)).strftime('%Y-%m-%d'))
```

```python
df_technology
```

```python
def findStockChange(row, dataset = df_msft, dateOffset = 0):
    currentstockDay = None
    date = datetime.strptime(row, '%Y-%m-%d')
    date = date + dt.timedelta(days=dateOffset)
    row = date.strftime('%Y-%m-%d')
    currentstockDay = dataset[dataset['Date'] == row]
    if not currentstockDay.empty:
        return currentstockDay.iloc[0]['Close'] > currentstockDay.iloc[0]['Open']
    else:
        return False

    
df_technology['ms_today'] = df_technology['TIMESTAMP'].apply(lambda row: findStockChange(row)).astype(np.int32)
```

```python
df_technology
```

```python
df_technology['ms_tomorrow'] = df_technology['TIMESTAMP'].apply(lambda row: findStockChange(row, dateOffset=1)).astype(np.int32)
```

```python
def normalize_headline(row):
    result = row.lower()
    #Delete useless character strings
    result = result.replace('...', '')
    whitelist = set('abcdefghijklmnopqrstuvwxyz 0123456789.,;\'-:?')
    result = ''.join(filter(whitelist.__contains__, result))
    return result

df_technology['normalized_headline'] = df_technology['TITLE'].apply(normalize_headline)


    
```
